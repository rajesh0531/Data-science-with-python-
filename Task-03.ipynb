```python
import matplotlib.pyplot as plt

# Bar Chart: Average GPA by GradeClass
avg_gpa_by_gradeclass = df.groupby('GradeClass')['GPA'].mean()

plt.figure(figsize=(10, 6))
avg_gpa_by_gradeclass.plot(kind='bar', color='skyblue')
plt.title('Average GPA by Grade Class')
plt.xlabel('Grade Class')
plt.ylabel('Average GPA')
plt.xticks(rotation=0)
plt.legend(['Average GPA'])
plt.show()

# Line Chart: StudyTimeWeekly vs. GPA
plt.figure(figsize=(10, 6))
plt.plot(df['StudyTimeWeekly'], df['GPA'], 'o-', color='green')
plt.title('Study Time Weekly vs. GPA')
plt.xlabel('Study Time Weekly (hours)')
plt.ylabel('GPA')
plt.legend(['GPA'])
plt.grid(True)
plt.show()
```
